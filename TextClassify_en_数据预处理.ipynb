```python
# 1. HTML提取文本

def remove_html(text):
    soup = BeautifulSoup(text, 'lxml')
    html_free = soup.get_text()
    return html_free
```

```python
# 2. 移除标点

def remove_punctuation(text):
    no_punct = ''.join([c for c in text if c not in string.punctuation])
    return no_punct
```

```python
# 3. 分词

# Some other examples of RegEx are:
# ‘\w+|\$[\d\.]+|\S+’ = splits up by spaces or by periods that are not attached to a digit
# ‘[A-Z]\w+’ = only words that begin with a capital letter.

tokenizer = RegexpTokenizer(r'\w+')
```

```python
# 4.移除停用词

def remove_stopwords(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words
```

```python
# 5. 词干提取
# nltk.stemmer.porter.PorterStemmer 类是一个用于从英文单词中获得符合语法的（前缀）词干的极其便利的工具。

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text

def word_stemmer(text):
    stem_text = ' '.join([stemmer.stem(i) for i in text])
    return stem_text
```
